```python
import pandas as pd
from sklearn.datasets import load_iris
iris = load_iris()
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
sepal_petal_df = df[['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)']]
display(sepal_petal_df.head())
```

```python
import matplotlib.pyplot as plt
x = sepal_petal_df['sepal length (cm)']
y = sepal_petal_df['sepal width (cm)']
plt.scatter(x, y)
plt.xlabel('sepal length (cm)')
plt.ylabel('sepal width (cm)')
plt.title('Scatter Plot of Sepal Length vs. Sepal Width')
plt.show()
```
